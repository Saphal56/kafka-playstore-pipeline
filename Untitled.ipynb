```python
# List of columns to drop
columns_to_drop = [
    'App Id',
    'Developer Id',
    'Developer Website',
    'Developer Email',
    'Privacy Policy',
    'Scraped Time',
    'Currency'
]

# Drop the columns
df_cleaned = df.drop(columns=columns_to_drop, errors='ignore')
```

```python
df_cleaned = df_cleaned.drop_duplicates()
print("After removing full duplicates:", df_cleaned.shape)
```
